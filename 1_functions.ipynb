```python
# Cette fonction permet de vérifier l'unicité des valeurs d'une colonne ou celle des combinaisons de valeurs entre 
# plusieurs colonnes. Elle donne le nombre de doublons (en comptant les premières occurrences), et indique le 
# nombre de ces doublons qui sont des premières occurrences.

# Par défaut, l'argument 'columns' a pour valeur 'all', ce qui indique à la fonction de calculer le nombre de 
# lignes qui sont des doublons sur l'ensemble des colonnes. On peut aussi entrer dans l'argument 'columns', soit un 
# soit plusieurs en-têtes de colonnes, lesquels doivent être des chaînes de caractères. Si l'on entre plusieurs 
# en-têtes, ceux-ci doivent être compris dans une liste. Dans ce dernier cas, la fonction vérifie si des lignes 
# sont des doublons sur l'ensemble des colonnes indiquées.

# Dans l'argument 'rows' il s'agit d'indiquer le nombre de lignes avec doublons que l'on veut afficher 
# (à supposer qu'il y en ait) :
# - mettre 'all' pour afficher toutes les lignes
# - mettre l'entier de son choix pour que s'affiche ce nombre de lignes (ou du moins toutes les lignes avec 
#   doublons s'il n'y en a pas autant que le nombre indiqué)

# L'argument 'display_all_columns' permet d'indiquer, en entrant la valeur True, que toutes les colonnes du 
# dataframe doivent être affichées. Si l'on ne le fait pas, il faut laisser la valeur par défaut (False), et le 
# nombre de colonnes affichées dépendra du nombre de colonnes du dataframe et des réglages d'affichage en vigueur.

# Cette fonction requiert l'importation suivante : from IPython.display import display





def duplicates(df, columns='all', rows=0, display_all_columns=False):
    
    df_name = [x for x in globals() if globals()[x] is df][0]
    
    if columns=='all':
        all_duplicates = df.loc[df.duplicated(keep=False)]
        if len(all_duplicates) == 0:
            print(f"\nLe dataframe {df_name} ne contient aucun doublon.")
        else:
            if (type(rows)==int) | (rows=='all'):
                print(f"\nLe dataframe {df_name} contient {len(all_duplicates)} doublons \
(en comptant les premières occurrences, au nombre de {len(all_duplicates) - sum(df.duplicated())}).")
                if (type(rows)==int) & (rows>0) & (rows<len(all_duplicates)):
                    print(f"Échantillon de {rows} de ces lignes doublons :")
                elif ((type(rows)==int) & (rows>=len(all_duplicates))) | (rows=='all'):    
                    print(f"Voici ces lignes doublons :")
                if display_all_columns==False:
                    with pd.option_context('display.max_rows', rows, 'display.min_rows', rows):
                        display(df[df.duplicated(keep=False)].sort_values(list(df.columns)))
                elif display_all_columns==True:
                    with pd.option_context('display.max_rows', rows, 'display.min_rows', rows, 
                                           'display.max_columns', None):
                        display(df[df.duplicated(keep=False)].sort_values(list(df.columns)))

    elif (type(columns)==str) | ((type(columns)==list) & (len(columns)==1) & (type(columns[0])==str)):
        if sum(df.duplicated(subset=columns)) == 0:
            print(f"\nLa colonne {columns} du dataframe {df_name} ne contient aucun doublon.")
        else:
            if (type(rows)==int) | (rows=='all'):
                if (type(rows)==int) & (rows==0):
                    print(f"\nLa colonne {columns} du dataframe {df_name} contient \
{sum(df.duplicated(subset=columns, keep=False))} doublons (en comptant les premières occurrences, au nombre de \
{sum(df.duplicated(subset=columns, keep=False)) - sum(df.duplicated(subset=columns))}).")
                elif (type(rows)==int) & (rows>0) & (rows<sum(df.duplicated(subset=columns, keep=False))):
                    print(f"\nLa colonne {columns} du dataframe {df_name} contient \
{sum(df.duplicated(subset=columns, keep=False))} doublons (en comptant les premières occurrences, au nombre de \
{sum(df.duplicated(subset=columns, keep=False)) - sum(df.duplicated(subset=columns))}). \
Échantillon de {rows} des lignes où se trouvent ces doublons :")
                    if display_all_columns==False:
                        with pd.option_context('display.max_rows', rows, 'display.min_rows', rows):
                            display(df[df.duplicated(subset=columns, keep=False)].sort_values(columns))
                    elif display_all_columns==True:
                        with pd.option_context('display.max_rows', rows, 'display.min_rows', rows, 
                                               'display.max_columns', None):
                            display(df[df.duplicated(subset=columns, keep=False)].sort_values(columns))
                elif ((type(rows)==int) & (rows>=sum(df.duplicated(subset=columns, keep=False)))) | (rows=='all'):
                    print(f"\nLa colonne {columns} du dataframe {df_name} contient \
{sum(df.duplicated(subset=columns, keep=False))} doublons (en comptant les premières occurrences, au nombre de \
{sum(df.duplicated(subset=columns, keep=False)) - sum(df.duplicated(subset=columns))}). \
Ces doublons se trouvent aux lignes suivantes :")
                    if display_all_columns==False:
                        with pd.option_context('display.max_rows', None):
                            display(df[df.duplicated(subset=columns, keep=False)].sort_values(columns))
                    elif display_all_columns==True:
                        with pd.option_context('display.max_rows', None, 'display.max_columns', None):
                            display(df[df.duplicated(subset=columns, keep=False)].sort_values(columns))

    elif (type(columns)==list) & (len(columns)>1) & (all(isinstance(i, str) for i in columns)):
        if sum(df.duplicated(subset=columns)) == 0:
            print(f"\nLe bloc de colonnes {columns} dans le dataframe {df_name} ne contient aucun doublon.")
        else:
            if (type(rows)==int) | (rows=='all'):
                if (type(rows)==int) & (rows==0):
                    print(f"\nLe bloc de colonnes {columns} dans le dataframe {df_name} contient \
{sum(df.duplicated(subset=columns, keep=False))} doublons (en comptant les premières occurrences, au nombre de \
{sum(df.duplicated(subset=columns, keep=False)) - sum(df.duplicated(subset=columns))}).")
                elif (type(rows)==int) & (rows>0) & (rows<sum(df.duplicated(subset=columns, keep=False))):
                    print(f"\nLe bloc de colonnes {columns} dans le dataframe {df_name} contient \
{sum(df.duplicated(subset=columns, keep=False))} doublons (en comptant les premières occurrences, au nombre de \
{sum(df.duplicated(subset=columns, keep=False)) - sum(df.duplicated(subset=columns))}). \
Échantillon de {rows} des lignes où se trouvent ces doublons :")
                    if display_all_columns==False:
                        with pd.option_context('display.max_rows', rows, 'display.min_rows', rows):
                            display(df[df.duplicated(subset=columns, keep=False)].sort_values(columns))
                    if display_all_columns==True:
                        with pd.option_context('display.max_rows', rows, 'display.min_rows', rows, 
                                               'display.max_columns', None):
                            display(df[df.duplicated(subset=columns, keep=False)].sort_values(columns))
                elif ((type(rows)==int) & (rows>=sum(df.duplicated(subset=columns, keep=False)))) | (rows=='all'):
                    print(f"\nLe bloc de colonnes {columns} dans le dataframe {df_name} contient \
{sum(df.duplicated(subset=columns, keep=False))} doublons (en comptant les premières occurrences, au nombre de \
{sum(df.duplicated(subset=columns, keep=False)) - sum(df.duplicated(subset=columns))}). \
Ces doublons se trouvent aux lignes suivantes :")
                    if display_all_columns==False:
                        with pd.option_context('display.max_rows', None):
                            display(df[df.duplicated(subset=columns, keep=False)].sort_values(columns))
                    if display_all_columns==True:
                        with pd.option_context('display.max_rows', None, 'display.max_columns', None):
                            display(df[df.duplicated(subset=columns, keep=False)].sort_values(columns))
```
